# Energy ANN

## Importing labraries

```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

## Data Preprocessing

```python
dataset = pd.read_excel('Folds5x2_pp.xlsx')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
print(y)
```

### Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 0)
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

## Building the model

```python
ann = tf.keras.models.Sequential() ##initializing the ANN model

## Adding the hidden layers
ann.add(tf.keras.layers.Dense(units = 6, activation = 'relu'))
ann.add(tf.keras.layers.Dense(units = 6, activation = 'relu'))

## Adding the output layer
ann.add(tf.keras.layers.Dense(units = 1))

##binary classification: activation = sigmoid
##no binary = softmax
#regression = no activation function
```

## Training the model

```python
ann.compile(optimizer = 'adam',loss = 'mean_squared_error')
```

```python
ann.fit(X_train, y_train, batch_size = 32, epochs = 100)
```

## Predicting

```python
y_pred = ann.predict(X_test)
```

```python

np.set_printoptions(precision = 2, threshold = 10000)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_pred),1)),axis = 1)) 
```

```python
final_accuracy = []
for i in range(0,len(y_pred)):
    final_accuracy.append(abs(y_pred[i]-y_test[i]))
```

```python
from math import sqrt
mean = 0
std = 0

for i in range(0,len(y_pred)):
    mean+=final_accuracy[i] 
    
mean = mean/len(y_pred)

for i in range(0,len(y_pred)):
    std+=((final_accuracy[i]-mean)*(final_accuracy[i]-mean))/len(y_pred)
std = round(sqrt(std),2)
```

```python
print('The mean of the error rate is: ',mean)
print('The standard deviation of the error rate is: ',std)
```
